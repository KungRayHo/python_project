```python
# 把原先DateId(ex：20170105)轉成new_dataId(ex：2017-01-05)

df["new_dataId"]=pd.to_datetime(df["DateId"].astype(str),format="%Y%m%d")
```

```python
# 把相同日期的銷售量加總

data=df.groupby(by="new_dataId")["Quantity"].sum()
data.head(10)
```

```python
# visualization

%matplotlib inline
data.plot(figsize=(15,5))
```
