```python
def GeneralData(line,datalist1,datalinst2):
    temp=line.split(",")
    emotion=np_utils.to_categorical(temp[0],7)
    imgStr=temp[1].split(" ")
    img=np.array(imgStr, 'float32')
    datalist1.append(ImgPrepare(img))
    datalinst2.append(emotion)
    return datalist1,datalinst2
```

```python
def ImgPrepare(imgNpArray):
    imgNpArray=imgNpArray.reshape(48,48,1)
    imgNpArray=imgNpArray/255
    return imgNpArray
```

```python
with open("fer2013.csv") as f:
    x_Train,y_Train,x_Test,y_Test=[],[],[],[]
    for line in f: 
        if "Training" in line:
            x_Train,y_Train=GeneralData(line,x_Train,y_Train)
        elif "PublicTest" in line:
            x_Test,y_Test=GeneralData(line,x_Test,y_Test)
x_Train=np.array(x_Train)
y_Train=np.array(y_Train)
x_Test=np.array(x_Test)
y_Test=np.array(y_Test)
```

```python
model=Sequential()
model.add(Conv2D(filters=16,kernel_size=(5,5),padding="same",input_shape=(48,48,1),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(filters=36,kernel_size=(5,5),padding="same",activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(7,activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
train_history=model.fit(x=x_Train,y=y_Train,validation_split=0.2,epochs=20,batch_size=300,verbose=2)
```
