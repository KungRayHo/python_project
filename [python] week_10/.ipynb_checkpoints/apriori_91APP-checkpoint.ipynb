```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df=pd.read_csv("../NTU_DataSet/Ntu_Orders.csv")
df.head()
```

```python
df.info()
```

```python
data_interesting=df.groupby(['MemberId', 'LevelFourCategoryName'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('MemberId')
data_interesting.head()
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

data_interesting_sets = data_interesting.applymap(encode_units)
```

```python
frequent_itemsets = apriori(data_interesting_sets, min_support=0.07, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules=rules[rules["confidence"]>0.6]
rules
```
