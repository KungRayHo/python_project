消費者是否會繼續消費取決於第一次消費的單價和財力

## 1. load data

```python
import numpy as np
import pandas as pd
```

```python
# get data from csv
data_shop=pd.read_csv("Ntu_Orders.csv",low_memory=False,dtype=str)
data=pd.read_csv("NTU_1317_Member.txt",delimiter="\t",low_memory=False)
```

## 2. Preparing the data

```python
data_member=data.copy()
data_member=data_member.dropna(axis=1,how="any")
del data_member["ValidFlag"]
data_member
```

```python
data_member=data_member[data_member["Recency"]!=-1]
del data_member["ShopId"]
data_member
```

```python
data_member["Category"]=np.zeros(len(data_member["MemberId"]))
```

```python
np_Quantity=np.array(data_shop["Quantity"].astype(float))
np_Price=np.array(data_shop["UnitPrice"].astype(float))
data_shop["price"]=np_Quantity*np_Price
```

```python
dic_price=dict(data_shop.groupby("MemberId")["price"].sum())
dic_frequency=dict(data_shop["MemberId"].value_counts())
```

```python
temp_dic={"index":list(dic_price.keys()),"price":list(dic_price.values()),"frequency":list(dic_frequency.values())}
temp_pd=pd.DataFrame(temp_dic).set_index("index")
temp_pd
```

```python
temp=temp_pd[(temp_pd["frequency"]>=np.median(list(dic_frequency.values())))&(temp_pd["price"]>=np.median(list(dic_price.values())))]
category_4=zip(temp.index,[4]*len(temp))
temp=temp_pd[(temp_pd["frequency"]<np.median(list(dic_frequency.values())))&(temp_pd["price"]>=np.median(list(dic_price.values())))]
category_3=zip(temp.index,[3]*len(temp))
temp=temp_pd[(temp_pd["frequency"]>=np.median(list(dic_frequency.values())))&(temp_pd["price"]<np.median(list(dic_price.values())))]
category_2=zip(temp.index,[2]*len(temp))
temp=temp_pd[(temp_pd["frequency"]<np.median(list(dic_frequency.values())))&(temp_pd["price"]<np.median(list(dic_price.values())))]
category_1=zip(temp.index,[1]*len(temp))
category_dic=dict(list(category_1)+list(category_2)+list(category_3)+list(category_4))
```

```python
data_member["MemberId"]=data_member["MemberId"].map(category_dic).fillna(0)
```

```python
data_member["MemberId"]=data_member["MemberId"].astype(float)
data_member["Recency"]=data_member["Recency"].astype(float)
data_member["Frequency"]=data_member["Frequency"].astype(float)
data_member["Monetary"]=data_member["Monetary"].astype(float)
```

```python
data_member.info()
```

```python
 data_member["RegisterDateTime"]=pd.datetime( data_member["RegisterDateTime"])
```

```python
data_member["MinOrderDate"] = pd.to_datetime(data_member["MinOrderDate"]) 
data_member["MinOrderDate"]=(data_member["MinOrderDate"] - data_member["RegisterDateTime"].min())/np.timedelta64(1,'D')
```

```python
data_member["RegisterDateTime"] = pd.to_datetime(data_member["RegisterDateTime"]) 
data_member["RegisterDateTime"]=(data_member["RegisterDateTime"] - data_member["RegisterDateTime"].min())/np.timedelta64(1,'D')
```
