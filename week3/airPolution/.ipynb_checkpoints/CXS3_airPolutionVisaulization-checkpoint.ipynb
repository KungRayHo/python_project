```python
import pandas as pd
import os
import glob
import re
cwd=os.getcwd()
```

```python
os.chdir(cwd)
os.chdir("./data")
father_folder_list=os.listdir()
filePath_list=[]
placeName=[]
for i in father_folder_list:
    os.chdir("./"+i)
    for file in os.listdir():
        if ".xls" in file:
#             print("./data/"+i+"/"+file)
            placeName.append(file[file.find("年")+1:file.find("站")])
            filePath_list.append("./data/"+i+"/"+file)
    os.chdir("..")
os.chdir(cwd)
# print(filePath_list)
# len(placeName)
```
